```python
import numpy
import sys

step_counter = 0

def quicksort(array):
    global step_counter
    if len(array) < 2:
        return array
    pivot = array.pop()
    left_array = []
    right_array = []
    for element in array:
        # Step for adding element to left or right array pivot
        step_counter += 1
        if element <= pivot:
            left_array.append(element)
        else:
            right_array.append(element)
    array.append(pivot)
    # Steps for the recursive calls
    step_counter += 2
    return numpy.concatenate((quicksort(left_array), numpy.array([pivot], dtype=int), quicksort(right_array)))
```

```python

unsorted_arr = [5,2,4,6,1,3]
quicksort(unsorted_arr)
print(step_counter)
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import math
import random

mpl.rcParams['figure.figsize'] = [16,4]
#mpl.rcParams.update({'font.size': 12})

from IPython.display import clear_output
fig, ax = plt.subplots(1, 1)
fig.suptitle('Quicksort')


B=[5,2,4,6,1,3]
points=13 # plotting points
x=[]
y=[]
ref=[]
c=[]

def generate_random_array(length):
    return [random.randint(1, 100) for _ in range(length)]

for i in range(0,points):
    clear_output(wait=True)
    print(i+1,"/", points)
    global step_counter
    step_counter = 0
    length_of_b = len(B)
    B = generate_random_array(length_of_b)
    quicksort(B)
    print(B)
    steps=step_counter 
    # double the size of the list to sort
    B=B*2
    x.append(len(B))
    y.append(steps)    
    a=len(B)*math.log(len(B),2)
    c.append(steps/(a))# quick and dirty estimation of c--> T(sort)=c.log(N)
    
i = 0
for v in x:
    ref.append([v*math.log(v) *c[i]])
    #ref.append(v*math.log((v),2))
    i += 1
ax.plot(x,y, color="red", label="counted steps")
ax.plot(x,ref, color="green", label= str(round(c[6],2)) + ".log(n) --> O(n log(n))")
ax.set_xlabel("size of n")
ax.set_ylabel("computation steps")
ax.legend()

# ax[1].scatter(x,c, label= "approximation of c" )
# ax[1].set_xlabel("size of n")
# ax[1].set_ylabel("c factor")

print(x)
print(y)
    
    
```
